# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp500_history.csv`: Contains historical closing prices of the S&P 500 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
# set the Path using pathlib
whale_returns_data = Path("Resources/whale_returns.csv")

# read the CSV, set the "Date" as the initial index column; parse the Date; establish the infer datetime format
whale_daily_returns_dataframe = pd.read_csv(whale_returns_data, index_col='Date', infer_datetime_format=True, parse_dates=True)

# sort data by ascending order and setting the dataframe with inplace=True
whale_daily_returns_dataframe.sort_index(ascending=True, inplace=True)

# check/validate your data
whale_daily_returns_dataframe.head()
whale_daily_returns_dataframe.tail()
```

```python
# Count nulls
# check for the number of nulls
whale_daily_returns_dataframe.isnull().sum()
```

```python
# Drop nulls
whale_daily_returns_dataframe = whale_daily_returns_dataframe.dropna()

# Check to ensure nulls were removed 
whale_daily_returns_dataframe.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns, set the Path using pathlib
algo_returns_data = Path("Resources/algo_returns.csv")

# read the CSV, set the "Date" as the initial index column; parse the Date; establish the infer datetime format
algo_daily_returns_dataframe = pd.read_csv(algo_returns_data, index_col='Date', infer_datetime_format=True, parse_dates=True)

# sort data by ascending order and setting the dataframe with inplace=True
algo_daily_returns_dataframe.sort_index(ascending=True, inplace=True)

# check/validate your data, checking first 10 rows
algo_daily_returns_dataframe.head(10)
```

```python
# Count nulls
algo_daily_returns_dataframe.isnull().sum()
```

```python
# Drop nulls
algo_daily_returns_dataframe = algo_daily_returns_dataframe.dropna()

# Confirm nulls were dropped
algo_daily_returns_dataframe.isnull().sum()
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices, set the Path using pathlib
sp500_history_data = Path("Resources/sp500_history.csv")

# read the CSV, set the "Date" as the initial index column; parse the Date; establish the infer datetime format
sp500_history_dataframe = pd.read_csv(sp500_history_data, index_col='Date', infer_datetime_format=True, parse_dates=True)

# sort data by ascending order and setting the dataframe with inplace=True
sp500_history_dataframe.sort_index(ascending=True, inplace=True)

# check/validate your data, checking first 15 rows
sp500_history_dataframe.head(15)
```

```python
# Check Data Types
sp500_history_dataframe.dtypes

# sample the "Close" column
sp500_history_dataframe['Close'].head()
```

```python
# Fix Data Types, remove all the dollar signs and then covert the close column from the indicated "object" to a "float"
sp500_history_dataframe['Close'] = sp500_history_dataframe['Close'].str.replace('$', '')

# Check to see if dollar signs were removed and confirm data type
sp500_history_dataframe['Close'].head()
```

```python
# Calculate Daily Returns
# Convert dataype to float to be able to calculate numerical returns
sp500_history_dataframe['Close'] = sp500_history_dataframe['Close'].astype('float')

# Confirm that SP500 dataframe is a float
sp500_history_dataframe.dtypes

# Now Calculate Daily Returns(note - already sorted by ascending value)
sp500_daily_returns_df= sp500_history_dataframe.pct_change()

# validate daily return file, head and last
sp500_daily_returns_df.head()
```

```python
# Drop nulls
# identify the number of nulls
sp500_daily_returns_df.isnull().sum()

# drop the nulls
sp500_daily_returns_df.dropna(inplace=True)

#confirm nulls were dropped
sp500_daily_returns_df.isnull().sum()
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp500_daily_returns_df = sp500_daily_returns_df.rename(columns = {'Close': 'S&P500 Returns'})
sp500_daily_returns_df.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
combined_daily_returns_df = pd.concat([whale_daily_returns_dataframe, algo_daily_returns_dataframe, sp500_daily_returns_df], axis='columns', join='inner')
combined_daily_returns_df.tail()
```

```python
# Calculate  the daily standard deviation of S&P 500
# Determine which portfolios are riskier than the S&P 500

daily_standard_deviation.sort_values(ascending=False, inplace=True)
daily_standard_deviation

```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
volatility_annualized_std = daily_standard_deviation * np.sqrt(252)
volatility_annualized_std.sort_values(ascending=False, inplace=True)
volatility_annualized_std
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for all portfolios using a 21-day window
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P 500
3. Choose one portfolio, then calculate and plot the 60-day rolling beta between it and the S&P 500

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
Rolling_21Day_Daily_Returns_SMA = combined_daily_returns_df.rolling(window=21).std()
# Plot the rolling standard deviation
Rolling_21Day_Daily_Returns_SMA.plot(title="21 Day SMA Volatility", figsize=(20,10))
```

### Calculate and plot the correlation

```python
# Calculate the correlation
correlation_df = combined_daily_returns_df.corr()

# Display de correlation matrix
correlation_df
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio - Berkshire Hathaway, Inc 
covariance_brk = combined_daily_returns_df['BERKSHIRE HATHAWAY INC'].cov(combined_daily_returns_df['S&P500 Returns'])
covariance_brk

# Calculate variance of S&P 500
SP500_variance = combined_daily_returns_df['S&P500 Returns'].var()
SP500_variance
# Computing beta
brk_beta = covariance_brk / SP500_variance
brk_beta

# Plot beta trend# 
# Beta vs. correlation utilizing scatterplot from the Seaborn library and fitting a regression line

import seaborn as sns
sn.lmplot(x='S&P500 Returns', y='BERKSHIRE HATHAWAY INC', data=combined_daily_returns_df, aspect=1.5, fit_reg=True)
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half life for each portfolio, using standard deviation (`std`) as the metric of interest.


```python
# Use `ewm` to calculate the rolling window
EWA_21day = combined_daily_returns_df.ewm(halflife=21).std()
EWA_21day.plot(title = "21 Day EWM Avg", figsize=(20,10))
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratios = (combined_daily_returns_df.mean() * 252) / volatility_annualized_std
sharpe_ratios.sort_values(ascending=False, inplace=True)
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

## Re-run the risk analysis with your portfolio to see how it compares to the others

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized `std`
my_portfolio_daily_std_volatility = NPFCX_BALCX_WSHCX_portfolio_returns.std()
my_portfolio_annual_volatility = my_portfolio_daily_std_volatility * np.sqrt(252)
my_portfolio_annual_volatility

# Calculate the annualized std for the S&P500 and TSES&P during the same time frame as my portfolio (01/01/2020 - 06/06/2023)
SP500_TX60_daily_returns = SP500_TX60_combined_df.pct_change()
SP500_TX60_daily_std_volatility = SP500_TX60_daily_returns.std()
volatility_annual_SP500_TX60 = SP500_TX60_daily_std_volatility * np.sqrt(252)


print(f'My annual portfolio volatility is {my_portfolio_annual_volatility}\n')
print(f'As compared to the other volatility in the other portfolio :\n{volatility_annualized_std}\n')
print(f'However, as compared to the annual volatilty of the S&P500 and the TSE S&P60 during my same time frame:\n{volatility_annual_SP500_TX60}')
```

### Calculate and Plot Rolling 60-day Beta for Your Portfolio compared to the S&P 500

```python
# Calculate and plot Beta
rolling_60day_covariance_NPFCX = personal_port_daily_returns['NPFCX'].rolling(window=60).cov(personal_port_daily_returns['S&P500'])
rolling_60day_covariance_BALCX = personal_port_daily_returns['BALCX'].rolling(window=60).cov(personal_port_daily_returns['S&P500'])
rolling_60day_covariance_WSHCX = personal_port_daily_returns['WSHCX'].rolling(window=60).cov(personal_port_daily_returns['S&P500'])

# Calculate the rolling variance of S&P500 over 60 days
rolling_60day_SP500_variance = personal_port_daily_returns['S&P500'].rolling(window=60).var()

# Calculate 60-day rolling beta
rolling_beta_NPFCX = rolling_60day_covariance_NPFCX / rolling_60day_SP500_variance
rolling_beta_BALCX = rolling_60day_covariance_BALCX / rolling_60day_SP500_variance
rolling_beta_WSHCX = rolling_60day_covariance_WSHCX / rolling_60day_SP500_variance

# Set the figure and plot the datasets
ax = rolling_beta_NPFCX.plot(figsize=(20,10), title='Rolling 60-Day Beta')
rolling_beta_BALCX.plot(ax=ax)
rolling_beta_WSHCX.plot(ax=ax)

# Set the legend of the figure
ax.legend(["NPFCX", "BALCX", "WSHCX"])
```
